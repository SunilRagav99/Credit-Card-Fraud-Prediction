```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
X=df.iloc[ : ,:-1]
Y=df.iloc[: , -1]
X,Y=oversample.fit_resample(X,Y)
```

```python
X=pd.DataFrame(X)


Y=pd.DataFrame(Y)


names=['Time','V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20','V21','V22','V23','V24','V25','V26','V27','V28','Amount','Class']

data=pd.concat([X,Y],axis=1)

d=data.values

df=pd.DataFrame(d,columns=names)
display(df)

```

```python
print(df["Class"].value_counts())
```

```python
correlation_matrix = df.corr()

# Display correlation with 'Class'
print("Correlation with 'Class':")
print(correlation_matrix['Class'].sort_values(ascending=False))
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Initialize and fit SelectKBest
selector = SelectKBest(score_func=f_classif, k='all')
selector.fit(X, Y)

# Get scores and p-values
scores = selector.scores_
p_values = selector.pvalues_

# Create a DataFrame for feature scores
scores_df = pd.DataFrame({
    'Feature': X.columns,
    'Score': scores,
    'P-Value': p_values
})

# Sort by score
print("Feature scores:")
print(scores_df.sort_values(by='Score', ascending=False))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
# X = df.drop('Class', axis=1)
X=df.drop(["V23","V22","Amount","V25","V13","V8","V15","Class"],axis=1)
y = df.Class
# X = scalar.fit_transform(X)

```

```python
X
```

```python
X = scalar.fit_transform(X)
```

```python
X
```

```python
y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=23)  
X_pca = pca.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.4, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# Create a logistic regression model
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)
y_pred=model.predict(X_test)


# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Classification Report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
svc=SVC(kernel="rbf", C=1.0)
svc.fit(X_train,y_train)
y_predict_svc=svc.predict(X_test)
# Accuracy
accuracy = accuracy_score(y_test, y_predict_svc)
print(f"Accuracy: {accuracy:.2f}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_predict_svc)
print("Confusion Matrix:")
print(conf_matrix)

# Classification Report
class_report = classification_report(y_test, y_predict_svc)
print("Classification Report:")
print(class_report)
```
